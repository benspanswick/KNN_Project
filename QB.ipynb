```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KDTree

# Using data from: https://www.kaggle.com/speckledpingu/nfl-qb-stats/home
```

```python
df = pd.read_csv("C:/Users/btspanswick/Desktop/PYproject/data/QBStats.csv")
df['freq'] = df.groupby('qb')['qb'].transform('count')
df.head()
```

```python
grouped_df = df.drop(columns=['year', 'home_away'])
grouped_df = grouped_df.groupby(['qb']).mean()
grouped_df
```

```python
grouped_df.dtypes
```

```python
bigben = grouped_df.loc['Ben RoethlisbergerBRoethlisberger']
bigben
```

```python
qbs = grouped_df.values
ben = bigben.values
ben = ben.reshape(1, -1)

ben
```

```python
qbs_sarray = grouped_df.to_records(index=True)
```

```python
qbs_sarray
reference_df = pd.DataFrame(qbs_sarray)
```

```python
KDTree.valid_metrics
```

```python
treel1 = KDTree(qbs, metric='euclidean')                                
distl1, indl1 = treel1.query(ben, k=11)
print(indl1)  # indices of 2 closest neighbors
print(distl1)  # distances to 2 closest neighbors
```

```python
results = indl1.tolist()
print(results)
```

```python
most_similar = pd.DataFrame(reference_df.loc[results[0]])

print(most_similar)

```
